```python
#Splitting the Dataset into Dependent and Independent Variables 
X = df.drop(columns=['class'])
y = df['class'] 
```

```python
X.shape 
```

```python
y.isnull().sum()
```

```python
#Checking for Null Values in X
X.isnull().sum()
```

```python
#Importing the Libraries for Data Preprocessing 
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer 
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler 
```

```python
#Creating objects for the Classes 
label_enc = LabelEncoder()
y_encoded = label_enc.fit_transform(y)
```

```python
y_encoded
```

```python
#separating columns for One HOt Encoding and Ordinal Encoding 
OneHot = ['cap-surface','bruises','gill-attachment','gill-spacing','gill-size','stalk-shape','veil-type','ring-number']
ordinal = ['cap-shape','cap-color','odor','gill-color','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-color','ring-type','spore-print-color','population','habitat']
```

```python
#Applying One Hot Encoding and Ordinal Encoding with the help of Columns Transfer 
preprocessor = ColumnTransformer(
    transformers = [('onehot',OneHotEncoder(handle_unknown='ignore'),OneHot),
    ('ordinal',OrdinalEncoder(),ordinal)],remainder='drop')
```

```python
#Applying Pipeline to Ensure proper Preprocessing of the dataset
pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('scaler',StandardScaler(with_mean=False))])
```

```python
X_processed = pipeline.fit_transform(X)
```

```python
#Shape of X is changed from (8124,22) -> (8124,32)
X_processed.shape
```

```python
y_encoded.shape
```

```python
#Drawing a Graph to Check Biasing in the Dataset 
sns.countplot(x='class',data=df)
plt.title("Class Distribution : edible(0) vs Poisnous (1)")
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
#Countplot for Comparison for Gill Color of Different classes 
sns.countplot(x='gill-color',hue ='class',data=df)
plt.title('Gill color vs Class')
plt.xlabel('Gill color(encoded)')
plt.ylabel('Count')
plt.legend(title='Class')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Splitting the Dataset for Training and Testing 
X_train,X_test,y_train,y_test=train_test_split(X_processed,y_encoded,test_size=0.2,random_state=42,stratify=y)
```

```python
#Importing the required library for Algorithm
from sklearn.linear_model import LogisticRegression
```

```python
#Creating the object for algorithm 
model = LogisticRegression()
```

```python
#fitting the dataset in the Logistic Regression model
model.fit(X_train,y_train)
```

```python
#Predicting the dataset on test set 
y_pred =model.predict(X_test)
```

```python
y_pred
```

```python
#importing the libraries for metric evaluation
from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix
```

```python
score = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
classification_report = classification_report(y_test,y_pred)
matrix = confusion_matrix(y_test,y_pred)
```

```python
#Printing the Different Metrics 
print(score)
print(precision)
print(classification_report)
print(confusion_matrix)
```

```python
#Confusion Matrix for the Predicited Values
plt.figure(figsize=(6,4))
sns.heatmap(matrix,annot=True,fmt='d',cmap='Blues',xticklabels=['Edible','Poisonous'],yticklabels=['Edible','Poisonous'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
#comparing the actual predicted and predicted values
compare_df = pd.DataFrame({"Actual":y_test , 'Predicted':y_pred})
```

```python
#countplot for actual v/s Predicted class counts 
sns.countplot(data=compare_df,x="Actual",hue='Predicted',palette='pastel')
plt.title("Actual & Predicted Class Counts")
plt.xlabel("Actual Class")
plt.ylabel("Count")
plt.legend(title="Predicted")
plt.show()
```

```python
compare_df['Match'] = compare_df["Actual"] == compare_df["Predicted"]
```

```python
#Countplot for Correct v/s Incorrect predictions 
sns.countplot(x='Actual',hue='Match',data=compare_df,palette={True: 'green' , False:'red'})
plt.title("Correct vs Incorrect Prediction by Class")
plt.xlabel("Actual Class")
plt.ylabel("Count")
plt.legend(title='Correct Prediction')
plt.show()
```
